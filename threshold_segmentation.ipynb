```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

```python
image=plt.imread('/home/deepesh/Downloads/akshay.jpg')
print(image.shape)
plt.imshow(image)
```

```python
gray_image=rgb2gray(image)
plt.imshow(gray_image)
```

```python
gray_image.shape
```

```python
gray=gray_image.reshape(gray_image.shape[0]*gray_image.shape[1])
print(gray)
```

```python
mean1=gray.mean()
for i in range(len(gray)):
    if gray[i]>mean1:
        gray[i]=1
    else:
        gray[i]=0
gray=gray.reshape(gray_image.shape[0],gray_image.shape[1])
print(gray)
plt.imshow(gray)
```

```python
mean1=gray.mean()
for i in range(len(gray)):
    if gray[i]>mean1:
        gray[i]=3
    elif gray[i]>1:
        gray[i]=2
    elif gray[i]>0.5:
        gray[i]=1
    elif gray[i]>0.25:
        gray[i]=0.5
    elif gray[i]>0.15:
        gray[i]=0.25
    else:
        gray[i]=0
gray=gray.reshape(gray_image.shape[0],gray_image.shape[1])
print(gray)
plt.imshow(gray)
```
