```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

```python
image = plt.imread('/home/deepesh/Downloads/akshay.jpg')
plt.imshow(image)
```

```python
gray_image=rgb2gray(image)
sobel_horizontal = np.array([np.array([1, 2, 1]), np.array([0, 0, 0]), np.array([-1, -2, -1])])
print(sobel_horizontal, 'is a kernel for detecting horizontal edges')
 
sobel_vertical = np.array([np.array([-1, 0, 1]), np.array([-2, 0, 2]), np.array([-1, 0, 1])])
print(sobel_vertical, 'is a kernel for detecting vertical edges')
```

```python
horizontal=ndimage.convolve(gray_image,sobel_horizontal,mode='reflect')
vertical=ndimage.convolve(gray_image,sobel_vertical,mode='reflect')
```

```python
plt.imshow(horizontal)
```

```python
plt.imshow(vertical)
```

```python
kernel_laplace = np.array([np.array([1, 1, 1]), np.array([1, -8, 1]), np.array([1, 1, 1])])
print(kernel_laplace)
```

```python
edge=ndimage.convolve(gray_image,kernel_laplace,mode='reflect')
plt.imshow(edge)
```
